```python
import itertools
distances = [[0, 10, 11, 8, 16, 20],
            [10, 0, 5, 4, 9, 6],
            [11, 5, 0, 7, 11, 14],
            [8, 4, 7, 0, 6, 15],
            [16, 9, 11, 6, 0, 7],
            [20, 6, 14, 15, 7, 0]]

cities = ['0', '1', '2', '3', '4', '5']

def obtain_distance(city1, city2):
    #obtain distance between two given cities
    return distances[int(city1)][int(city2)]

#stores shortest distance to home through route in key
#key is tuple with starting city, list of cities. value is a numerical value.
memo = dict()
```

```python
def iterativeHeldKarp(cities):
    
    #consider all possible sizes of problems from 1 city to 6 cities
    for subset_size in range(1,len(cities)+1):

        #for a given size, consider all the possible subproblems
        for subproblem in itertools.permutations(cities, subset_size):
            #initialize min distance to infinity
            min_distance = float('inf')

            #take the first city in the problem and calculate the distance to each possible
            #next city and the cost function for the next city and the intermediate cities on the way to home
            first_city = subproblem[0]
            cities_to_visit = [city for city in subproblem if city != first_city and city != '0']
            
            if not cities_to_visit:
                test_subproblem_distance = obtain_distance(first_city,'0')
                min_distance = min(test_subproblem_distance, min_distance)
                memo[(first_city, frozenset([]))] = min_distance
            else:
                for next_city in cities_to_visit:
                    test_distance = obtain_distance(first_city, next_city)
                    next_cities_to_visit = [city for city in cities_to_visit if city != next_city and city != '0']
                    inner_dist = memo[(next_city, frozenset(next_cities_to_visit))]
                    test_subproblem_distance = test_distance + inner_dist
                    min_distance = min(test_subproblem_distance, min_distance)

                memo[(first_city, frozenset(cities_to_visit))] = min_distance
              
            min_distance = min(test_subproblem_distance, min_distance) 

    #return the optimal tour starting at home, visiting all other cities, and returning home
    return memo[('0', frozenset(['1','2','3','4','5']))]
```

```python
iterativeHeldKarp(cities)
```
